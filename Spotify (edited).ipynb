```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import time

# Replace these with your actual Spotify API credentials
client_id = 'Yr_Client_id'
client_secret = 'Yr_client_secret'

# Authenticate with the Spotify API
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))

# Function to search for track ID
def get_track_id(row):
    query = f"track:{row['track_name']} artist:{row['artist(s)_name']}"
    try:
        results = sp.search(q=query, type='track', limit=1)
        if results['tracks']['items']:
            return results['tracks']['items'][0]['id']
        else:
            return None
    except Exception as e:
        print(f"Error fetching track ID for {row['track_name']} by {row['artist(s)_name']}: {e}")
        return None

# Function to get album cover URL
def get_album_cover_url(track_id):
    try:
        track_info = sp.track(track_id)
        return track_info['album']['images'][0]['url'] if track_info['album']['images'] else None
    except Exception as e:
        print(f"Error fetching album cover URL for track ID {track_id}: {e}")
        return None

# Add track_id and album_cover_url columns to the dataset
df['track_id'] = df.apply(get_track_id, axis=1)
time.sleep(1)  # To avoid hitting rate limits
df['album_cover_url'] = df['track_id'].apply(get_album_cover_url)

# Save the updated dataset
df.to_csv('spotify_updated.csv', index=False)
```
